# Goldstein 3e — Chapter 1, Problem 11

> Decide whether **F** = yz **i** + zx **j** + xy **k** is conservative.

**Book reference:** Exercises, p. 31 (PDF p. 50)

---

## 1. Physical setup

Force components: $F_x = yz$, $F_y = zx$, $F_z = xy$.

**Question:** Is there a potential $V$ such that **F** = −grad $V$?

## 2. Intuition

Conservative means path-independent work — zero net work around any closed loop.

Equivalent test (simply connected region): curl **F** = **0**.

## 3. Work-integral test (F · ds)

**Line element:** $d\mathbf{s} = dx\,\mathbf{i} + dy\,\mathbf{j} + dz\,\mathbf{k}$

**Dot product:**

$$\mathbf{F}\cdot d\mathbf{s} = yz\,dx + zx\,dy + xy\,dz$$

**Exact differential check:** with $V = -xyz$,

$$-dV = d(xyz) = yz\,dx + xz\,dy + xy\,dz = \mathbf{F}\cdot d\mathbf{s}$$

So the integrand is a perfect differential → path-independent.

**Path integration** from $(a,b,c)$ to $(x,y,z)$ via axis-aligned segments:

$$W = bc(x-a) + xc(y-b) + xy(z-c) = xyz - abc$$

**Closed loop** (start = end): $W = 0$.

## 4. Cross-check — curl

$$(\nabla\times\mathbf{F})_x = \partial_y(xy) - \partial_z(zx) = x - x = 0$$

$$(\nabla\times\mathbf{F})_y = \partial_z(yz) - \partial_x(xy) = y - y = 0$$

$$(\nabla\times\mathbf{F})_z = \partial_x(zx) - \partial_y(yz) = z - z = 0$$

Potential: $V = -xyz + C$.

## 5. Takeaway

**Yes — conservative.** Work integral is exact; $V = -xyz$ (up to a constant).
