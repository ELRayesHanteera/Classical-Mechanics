> Find the orbital period and orbital angular velocity of a satellite in circular orbit at altitude 720 km above Earth.

**Given:** Earth radius \(R = 6000\) km, surface gravity \(g = 9.83\) m/s\(^2\).

**Book reference:** Exercises, p. 31 (PDF p. 50)

---

## 1. Physical setup

- Satellite in a **circular orbit** at altitude \(h = 720\) km.
- Orbital radius (from Earth's center): \(r = R + h\).
- Earth's gravity provides the centripetal force.
- Neglect atmosphere, Earth's rotation, and treat Earth as a point mass.

**Known constants**

| Symbol | Value |
|--------|-------|
| \(R\) | \(6000\) km \(= 6.00 \times 10^6\) m |
| \(h\) | \(720\) km \(= 7.20 \times 10^5\) m |
| \(r = R+h\) | \(6720\) km \(= 6.72 \times 10^6\) m |
| \(g\) | \(9.83\) m/s\(^2\) (at surface) |

## 2. Intuition

A satellite stays in orbit because gravity bends its path into a circle. At radius \(r\), the required centripetal acceleration is \(v^2/r\). Gravity supplies \(GM/r^2\), so

\[
\frac{v^2}{r} = \frac{GM}{r^2} \quad\Rightarrow\quad v = \sqrt{\frac{GM}{r}}.
\]

Once we know \(v\), one full orbit is just distance \(2\pi r\) at speed \(v\):

\[
T = \frac{2\pi r}{v}, \qquad \omega = \frac{v}{r} = \frac{2\pi}{T}.
\]

We are not given \(GM\) directly, but we **are** given \(g\) at the surface. Since \(g = GM/R^2\), we can eliminate \(GM\) and express everything in \(R\), \(r\), and \(g\).

## 3. Formal solution

### Step 1 â€” Circular orbit speed

Newton's second law for uniform circular motion:

\[
\frac{GMm}{r^2} = \frac{mv^2}{r} \quad\Rightarrow\quad v^2 = \frac{GM}{r}.
\]

Use \(GM = gR^2\):

\[
v^2 = \frac{gR^2}{r} \quad\Rightarrow\quad \boxed{v = R\sqrt{\frac{g}{r}} = R\sqrt{\frac{g}{R+h}}.}
\]

### Step 2 â€” Orbital period

\[
T = \frac{2\pi r}{v} = \frac{2\pi r}{R}\sqrt{\frac{r}{g}}
= \boxed{\frac{2\pi}{R\sqrt{g}}\, r^{3/2}
= 2\pi\sqrt{\frac{r^3}{gR^2}}.}
\]

(This is Kepler's third law for circular orbits: \(T^2 \propto r^3\).)

### Step 3 â€” Orbital angular velocity

\[
\omega = \frac{v}{r} = \sqrt{\frac{GM}{r^3}} = \boxed{\sqrt{\frac{gR^2}{r^3}}
= \frac{R}{r}\sqrt{\frac{g}{r}}.}
\]

Equivalently, \(\omega = 2\pi/T\).

### Step 4 â€” Numbers

\[
v = 6.00\times10^6 \sqrt{\frac{9.83}{6.72\times10^6}}
\approx 7.26\ \text{km/s}
\]

\[
T = \frac{2\pi (6.72\times10^6)}{7.26\times10^3}
\approx 5.82\times10^3\ \text{s} \approx 97.0\ \text{min}
\]

\[
\omega = \frac{7.26\times10^3}{6.72\times10^6}
\approx 1.08\times10^{-3}\ \text{rad/s}
\]

## 4. Sanity checks

| Check | Result |
|-------|--------|
| **Dimensions** | \([v]=\) m/s, \([T]=\) s, \([\omega]=\) rad/s âœ“ |
| **Low orbit** | ISS-like orbits (\(r \sim 6700\) km) have \(T \sim 90\)â€“\(95\) min â€” our 97 min is in the right ballpark |
| **Surface limit** | As \(r \to R\), \(T \to 2\pi R / \sqrt{g} \approx 84\) min (too fast for a real surface orbit, but the formula is consistent) |
| **Energy** | \(v \ll c\) â€” non-relativistic mechanics is valid |

```python
import math

R = 6000e3      # m
h = 720e3       # m
g = 9.83        # m/s^2
r = R + h

v = R * math.sqrt(g / r)
T = 2 * math.pi * r / v
omega = v / r

print(f"Orbital radius r = {r/1e3:.1f} km")
print(f"Orbital speed    v = {v/1e3:.3f} km/s  ({v:.1f} m/s)")
print(f"Period           T = {T:.1f} s  ({T/60:.2f} min)")
print(f"Angular velocity Ï‰ = {omega:.6f} rad/s")
```

## 6. Takeaway

For a circular orbit in a \(1/r^2\) gravitational field, once you know \(g\) at the surface you can write everything as \(v = R\sqrt{g/r}\), \(T = 2\pi r/v\), and \(\omega = v/r\). The \(r^{3/2}\) scaling of the period is Kepler's third law in action.
