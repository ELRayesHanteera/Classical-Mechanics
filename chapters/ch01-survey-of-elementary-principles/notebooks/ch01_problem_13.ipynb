> Derive the rocket equation for vertical flight with decreasing mass; integrate for constant fuel burn rate; estimate how much fuel (relative to empty rocket weight) is needed to reach escape velocity.

**Book reference:** Exercises, p. 31 (PDF p. 50)

## 1. Physical setup

**The scene:** A rocket stands on the launch pad, then burns fuel and climbs straight up. As fuel leaves the nozzle, the rocket gets lighter ÔÇö but gravity keeps pulling down on whatever mass remains.

```
         ÔćĹ +y (up)
         |
      [ rocket ]  ÔćÉ mass m(t), velocity v(t)
         |
      ~~~~ exhaust  Ôćĺ leaves backward at speed v' relative to rocket
         |
      ÔöÇÔöÇÔöÇ ground ÔöÇÔöÇÔöÇ
         g Ôćô
```

**Variables (with units):**

| Symbol | Meaning | Units |
|--------|---------|-------|
| $m(t)$ | Rocket mass (including remaining fuel) | kg |
| $v(t)$ | Rocket velocity (upward positive) | m/s |
| $v'$ | Exhaust speed **relative to the rocket** (magnitude) | m/s |
| $g$ | Uniform gravitational acceleration | m/s$^2$ |
| $m_0$ | Initial mass (full fuel + empty rocket) | kg |
| $m_e$ | Empty rocket mass (no fuel) | kg |

**Assumptions:**
- Motion is strictly vertical.
- No air resistance.
- Gravity is uniform: $\mathbf{g} = -g\,\hat{\jmath}$.
- Exhaust is expelled backward along the rocket axis; $v'$ is the speed of gas **relative to the rocket**.

**Part C numerical data:** $v' = 2.1$ km/s, $|\dot m| = m_0/60$ per second, starts from rest, target speed $v_{\text{esc}} \approx 11.2$ km/s.

## 2. Intuition

### What is the system really doing?

Imagine standing on a skateboard and throwing bricks backward. Each throw pushes you forward (Newton's third law). A rocket does the same thing continuously ÔÇö except the "bricks" are hot exhaust gas, and you're also fighting gravity the whole time.

This is **not** a fixed-mass problem. You cannot simply write $F = ma$ with constant $m$. Mass leaves the system every second.

### What is Goldstein really asking?

1. **Part A:** Where does the rocket equation come from? Ôćĺ We need **momentum conservation** for a system whose mass changes.
2. **Part B:** If fuel burns at a steady rate, how does speed depend on remaining mass? Ôćĺ Integrate the equation of motion.
3. **Part C:** How absurd is the fuel requirement for escape velocity? Ôćĺ Plug in numbers and see.

### Why momentum, not energy?

Thrust is an **internal** rearrangement of momentum (rocket + exhaust). The exhaust carries momentum away; the rocket recoils. Energy is useful too, but momentum conservation tracks the thrust term directly. Goldstein is building the habit: **variable mass Ôćĺ momentum first**.

### Strategy (the story before the algebra)

1. Zoom in on an infinitesimal burn interval $dt$. Write "before" and "after" momenta.
2. Include gravity as an external impulse $-mg\,dt$.
3. Expand, drop tiny products ($dm\,dv$), get a differential equation.
4. For Part B, use constant $\dot m = -m_0/60$ and integrate $v(m)$.
5. For Part C, set burnout speed equal to escape velocity and solve for fuel mass.

### Sign convention ÔÇö read this before Part A

- Upward is **+**.
- As fuel burns, $m$ **decreases**, so $dm < 0$ in an interval $dt > 0$.
- Exhaust is expelled **backward** (downward relative to an ascending rocket), so in the lab frame its velocity is $v - v'$ (rocket speed minus exhaust speed relative to rocket).

**Common trap:** Writing $m\,dv/dt = +v'\,dm/dt$ with the wrong sign. Since $dm/dt < 0$, the thrust term $-v'\,dm/dt$ is **positive upward** ÔÇö as it must be.

## 3. Formal solution ÔÇö Part A

**Governing principle:** Conservation of linear momentum (with external gravitational impulse).

### Step 1 ÔÇö States before and after a burn $dt$

**Before** the burn at time $t$:
- Rocket mass: $m$
- Rocket velocity: $v$
- Total momentum: $p = mv$

**After** the burn at time $t + dt$:
- Rocket mass: $m + dm$ (with $dm < 0$ because fuel was lost)
- Rocket velocity: $v + dv$
- Mass ejected in this interval: $-dm > 0$ (positive amount of gas)
- Ejected gas velocity in the lab frame: $v - v'$
  - *Why $v - v'$?* Rocket moves up at $v$; gas moves backward at $v'$ **relative to rocket** Ôćĺ lab speed is $v - v'$.

**After** momentum:
$$p' = (m + dm)(v + dv) + (-dm)(v - v')$$

The second term is the momentum of the ejected gas (mass $-dm$, velocity $v - v'$).

### Step 2 ÔÇö Impulse from gravity

Gravity is external. Upward positive, gravitational force is $-mg$, so impulse is:
$$J = (-mg)\,dt$$

Momentum theorem:
$$p' = p + J = mv - mg\,dt$$

### Step 3 ÔÇö Equate and expand

$$(m + dm)(v + dv) + (-dm)(v - v') = mv - mg\,dt$$

Expand the rocket term:
$$(m + dm)(v + dv) = mv + m\,dv + v\,dm + dm\,dv$$

Expand the exhaust term:
$$(-dm)(v - v') = -dm\,v + dm\,v'$$

Add them:
$$mv + m\,dv + v\,dm + \underbrace{dm\,dv}_{\text{2nd order}} - dm\,v + dm\,v' = mv - mg\,dt$$

The $v\,dm$ and $-dm\,v$ terms **cancel** (this is important ÔÇö they are not "missing physics," they cancel by construction).

Drop $dm\,dv$ (second-order small):
$$m\,dv + v'\,dm = -mg\,dt$$

*Plain language:* The rocket's momentum change $m\,dv$ plus the momentum carried away by exhaust ($v'\,dm$; note $dm<0$ so this term can push the rocket forward) equals the momentum lost to gravity.

### Step 4 ÔÇö Divide by $dt$

$$\boxed{m\frac{dv}{dt} = -v'\frac{dm}{dt} - mg}$$

**Reading the terms:**
- $-v'\,dm/dt$: **thrust**. Since $dm/dt < 0$, this term is **positive upward**.
- $-mg$: **weight**, always pulls down.

**What if $g = 0$?** We get $m\,dv/dt = -v'\,dm/dt$, the standard rocket equation in free space ÔÇö good limiting case.

## 3. Formal solution ÔÇö Part B

**Given:** constant mass-loss rate (in time, not as a fraction of current mass):
$$\frac{dm}{dt} = -\frac{m_0}{60}$$

Define $\mu \equiv -\dot m = m_0/60 > 0$ (kilograms per second of fuel consumed).

### Step 1 ÔÇö Switch from $t$ to $m$ as the independent variable

We want $v(m)$. Use the chain rule:
$$\frac{dv}{dt} = \frac{dv}{dm}\frac{dm}{dt}$$

Substitute into the rocket equation:
$$m\,\frac{dv}{dm}\,\frac{dm}{dt} = v'\,\frac{dm}{dt} - mg$$

Since $\dot m = -\mu \neq 0$, divide both sides by $dm/dt$:
$$m\,\frac{dv}{dm} = v' - \frac{g}{\mu}\,m$$

*Plain language:* At each mass $m$, the "acceleration per unit mass change" balances thrust ($v'$) against a gravity term proportional to current mass.

### Step 2 ÔÇö Integrate

Separate variables:
$$dv = v'\,\frac{dm}{m} - \frac{g}{\mu}\,dm$$

Integrate from initial state $(m_0, v_0)$ to generic $(m, v)$:
$$\int_{v_0}^{v} dv' = v'\int_{m_0}^{m}\frac{dm'}{m'} - \frac{g}{\mu}\int_{m_0}^{m} dm'$$

$$v - v_0 = v'\ln\!\left(\frac{m}{m_0}\right) - \frac{g}{\mu}(m - m_0)$$

Because $m < m_0$ during burn, $\ln(m/m_0) < 0$ and $(m - m_0) < 0$.

With $v_0 = 0$ and rearranging signs:
$$\boxed{v(m) = v'\ln\!\left(\frac{m_0}{m}\right) - \frac{g}{\mu}(m_0 - m)}$$

For this problem $\mu = m_0/60$:
$$\boxed{v(m) = v'\ln\!\left(\frac{m_0}{m}\right) - \frac{60g}{m_0}(m_0 - m)}$$

**Reading the two terms:**
1. $v'\ln(m_0/m)$ ÔÇö **thrust benefit**: logarithmic in mass ratio (Tsiolkovsky structure).
2. $-(60g/m_0)(m_0 - m)$ ÔÇö **gravity penalty**: linear in fuel used; gravity acts the entire burn.

**What if no fuel is burned ($m = m_0$)?** Both terms vanish Ôćĺ $v = 0$. Ôťô

## 3. Formal solution ÔÇö Part C

**Goal:** Find $W_{\text{fuel}}/W_{\text{empty}} = m_f/m_e$ needed so burnout speed reaches escape velocity.

### Step 1 ÔÇö Burnout condition

At burnout, remaining mass is $m = m_e$. Fuel mass burned:
$$m_f = m_0 - m_e$$

Let $r = m_f/m_e$ (fuel-to-empty **mass** ratio). Then:
$$m_0 = m_e + m_f = m_e(r + 1)$$

### Step 2 ÔÇö Burnout speed as a function of $r$

Substitute $m = m_e$ into $v(m)$:
$$v_{\text{burn}} = v'\ln\!\left(\frac{m_0}{m_e}\right) - \frac{60g}{m_0}(m_0 - m_e)$$

Use $m_0/m_e = r + 1$ and $(m_0 - m_e)/m_0 = r/(r+1)$:
$$\boxed{v_{\text{burn}} = v'\ln(r + 1) - \frac{60g\,r}{r + 1}}$$

### Step 3 ÔÇö Set $v_{\text{burn}} = v_{\text{esc}}$ and solve

Given: $v' = 2.1\times10^3$ m/s, $v_{\text{esc}} = 11.2\times10^3$ m/s, $g = 9.8$ m/s$^2$.

This transcendental equation is best solved numerically (see code cell below).

**Crude estimate** (when $m_0 \approx m_f$, i.e. empty rocket negligible):
$$v_{\text{esc}} \approx v'\ln\!\left(\frac{m_f}{m_e}\right) - 60g$$
$$\ln r \approx \frac{v_{\text{esc}} + 60g}{v'} = \frac{11200 + 588}{2100} \approx 5.61 \quad\Rightarrow\quad r \approx e^{5.61} \approx 274$$

Exact numerical solution:
$$\boxed{\frac{W_{\text{fuel}}}{W_{\text{empty}}} = \frac{m_f}{m_e} \approx 273}$$

Goldstein says **"almost 300"** ÔÇö same order of magnitude. The fuel mass must be **hundreds of times** the empty rocket. Most of the thrust goes into accelerating unburned fuel, while gravity acts for the entire ascent.

*Physical picture at burnout:* You need enough logarithmic thrust gain from shedding mass to overcome both escape-speed requirement **and** the cumulative gravity drag term $60gr/(r+1)$.

## 4. Sanity checks

| Check | Question | Result |
|-------|----------|--------|
| **Dimensions** | Is $v(m)$ in m/s? | $v'$ gives m/s; $(g/\mu)m$ has units $(m/s^2)(kg/(kg/s)) = m/s$ Ôťô |
| **No gravity** | Set $g=0$ | $v = v'\ln(m_0/m)$ ÔÇö classic Tsiolkovsky Ôťô |
| **No burn** | $m = m_0$ | $v = 0$ Ôťô |
| **Thrust sign** | $dm/dt < 0$ | $-v'\,dm/dt > 0$ upward Ôťô |
| **Order of magnitude** | $r \sim 300$ | Enormous but familiar from real rocketry ÔÇö most mass is fuel Ôťô |
| **Chapter link** | Variable mass in Ch. 1 | Momentum for systems of particles; preview of treating $m(t)$ carefully before Lagrangian methods |

````python
import math

# Part C numerical data
v_esc = 11.2e3      # escape speed (m/s)
v_prime = 2.1e3     # exhaust speed relative to rocket (m/s)
g = 9.8             # m/s^2

def v_burnout(r):
    """Burnout speed when fuel/empty mass ratio is r = m_f / m_e."""
    return v_prime * math.log(r + 1) - 60 * g * r / (r + 1)

# Solve v_burnout(r) = v_esc by bisection
lo, hi = 200.0, 350.0
for _ in range(60):
    mid = (lo + hi) / 2
    if v_burnout(mid) < v_esc:
        lo = mid
    else:
        hi = mid

r = (lo + hi) / 2

print(f"Fuel / empty mass ratio  m_f/m_e = {r:.1f}")
print(f"Burnout speed           v     = {v_burnout(r)/1e3:.4f} km/s")
print(f"Target escape speed              = {v_esc/1e3:.1f} km/s")

# Crude approximation for comparison
r_approx = math.exp((v_esc + 60 * g) / v_prime)
print(f"\nCrude estimate (m_0 Ôëł m_f):     r Ôëł {r_approx:.1f}")
````

## 6. Takeaway

**One sentence:** Variable-mass motion demands momentum conservation, not $F=ma$ with constant $m$; thrust appears as $-v'\dot m$, and reaching escape speed from rest with these parameters requires fuel weight **~270ÔÇô300├Ś** the empty rocket.

**Reusable habit:** Whenever mass leaves the system (rockets, chains, sand falling off conveyors), isolate a small interval, write before/after momenta, identify external impulses, then take the limit $dt \to 0$.

**Chapter connection:** Goldstein Ch. 1 treats systems of particles and external forces; this problem is the prototype for later handling constraints and generalized coordinates when mass is not conserved in the naive sense.
